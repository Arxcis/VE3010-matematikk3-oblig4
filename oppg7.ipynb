## Exercise 7c)

Plot the vector field on a suitable domain:

$$
\vec{F} = P\hat{i} + Q\hat{j} +  R\hat{k}
$$
$$
\vec{F} = \sin(x)\cos(x)\hat{i} + \cos(x + z)\hat{j} + \sin(y + z)\hat{k}
$$


```python
from numpy import meshgrid,arange,sin,cos
from matplotlib.pyplot import figure

x = arange(-1,1, 0.1)
y = arange(-1,1, 0.1)
z = arange(-1,1, 0.5)

x,y,z = meshgrid(x,y,z)

P = sin(x)*cos(x)
Q = cos(x+z)
R = sin(y+z)

figure(figsize=(6,6))\
    .add_subplot(projection="3d")\
    .quiver(x,y,z,P,Q,R,\
        color="blue",\
        length=0.2)
```
