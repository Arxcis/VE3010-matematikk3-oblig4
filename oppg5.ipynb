## Excercise 5a):

Plot level curves for the surface given by the following parametrisation:

$$
X(u,v) = \begin{cases}
    x=u& \text{for} -4 < u < 4\\
    y=v& \text{for} -4 < v < 4\\
    z=1-u^2-v^3-u+3v-uv
\end{cases}
$$

```python
from numpy import linspace, meshgrid,arange
from matplotlib.pyplot import figure

u = linspace(-4,4)
v = linspace(-4,4)
u,v = meshgrid(u,v)
z = 1 - u**2 - v**3 - u + 3*v - u*v

figure(figsize=(6,6))\
    .add_subplot()\
    .contour(u,v,z,\
        levels=arange(-20,20,1),\
        colors="blue")
```
