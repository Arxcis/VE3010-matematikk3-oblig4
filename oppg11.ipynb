Surface X given by:

$$

Vectorfields from exercise 7 given by:
$$
Fields = \begin{cases}
    a = (x+y)\hat{i}& + (x-yz)\hat{j}& + (xyz)\hat{k}\\
    b = (-xz)\hat{i}& - (xy)\hat{j}& + (e^{-xyz})\hat{k}\\
    c = (-\sin(x)\cos(y))\hat{i}& + (\cos(x+z))\hat{j}& + (\sin(y+z))\hat{k}\\

$$




```python
from numpy import linspace, meshgrid

x = linspace(-2,2)
y = linspace(-2,2)
x,y = meshgrid(x,y)
z = 4 - x**2 - y**2

from matplotlib.pyplot import figure

# Field exercise 7a)
fig = figure(figsize=(7,7))
ax = fig.add_subplot(projection="3d")
ax.plot_surface(x,y,z, cmap="plasma")

P = x + y
Q = x - y*z
R = x*y*z
ax.quiver(x,y,z,P,Q,R,\
        color="blue",\
        length=0.06)

fig = figure(figsize=(7,7))
fig.add_subplot()\
   .quiver(x,y,P,Q, color="blue", scale=15)

fig = figure(figsize=(7,7))
fig.add_subplot()\
   .quiver(y,z,Q,R, color="blue", scale=15)
```
