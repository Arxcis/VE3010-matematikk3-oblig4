Plot the following "monkey-saddle" surface X in $R^{3}$:
$$
    X(x,y) = (x,y,x^3 - 3xy^2);
$$

Plot in the domain U:
$$
    U = [-1,1] \times [-1,1]\\
$$

Parametric representation of "monkey-saddle:
$$
    f(t,u)=\begin{cases}
        x=u&  \text{for} -1 < u < 1\\
        y=v& \text{for} -1 < v < 1\\
        z=u^3-3uv^2
    \end{cases}
$$

```python
from numpy import linspace, meshgrid
from matplotlib.pyplot import figure

u = linspace(-1,1)
v = linspace(-1,1)
u,v = meshgrid(u,v)

z = u**3 - 3*u*v**2

figure(figsize = (6,6))\
    .add_subplot(projection="3d")\
    .plot_surface(u,v,z, cmap="plasma")
```
