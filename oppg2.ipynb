Parametrise and plot the following curve: the parabola which includes the points $(-1,1)$, $(0,0)$ and $(1,1)$.

Parabolic equation:
$$
    y=x^2\\
$$

Parametric representation of parabola:
$$
    f(t)=\begin{cases}
        x=t&\\ 
        y=t^2&
    \end{cases} 
    
    \text{for } -1 < t < 1
$$

```python
from numpy import linspace, pi, sin, cos
from matplotlib.pyplot import figure, plot

ax = figure(figsize = (6,6))\
    .add_subplot()

t = linspace(-1,1)
x = t
y = t**2

ax.plot(x,y, label="A parabola: y=x^2")
```
